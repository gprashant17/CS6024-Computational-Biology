```python
def FittingAlignment(v,w,n1,n2):
    path = {}
    DAG = {i:[0]*(n1+1) for i in range(n2+1)}
    for i in range(1,n2+1):
        # DAG[i][0] = 0
        for j in range(1,n1+1):
            if w[i-1] == v[j-1]:
                opt = [DAG[i-1][j-1]+1,DAG[i-1][j]-1,DAG[i][j-1]-1,0]
            else:
                opt = [DAG[i-1][j-1]-1,DAG[i-1][j]-1,DAG[i][j-1]-1,0]
            DAG[i][j] = max(opt)
#             print(DAG[i],end = " ")
            k = opt.index(max(opt))
            if k==0:
                path[(i,j)] = "m"
            elif k==1:
                path[(i,j)] = "i"
            elif k==2:
                path[(i,j)] = "d"
                
    print(DAG)
    print(path)
    last_row = DAG[n2]
    Score = max(last_row)
    # last_row.reverse()
    ind =  [i for i,val in enumerate(last_row) if val==Score][-1]
    pos = (n2,ind)
    str1 = ""
    str2 = ""
#     start = pos.copy()
#     print(pos)
    while pos[0]!=0:
        if path[pos] == "m":
            str1 = v[pos[1]-1] + str1
            str2 = w[pos[0]-1] + str2
            pos = (pos[0]-1,pos[1]-1)
        elif path[pos] == "i":
            str1 = "-" + str1
            str2 = w[pos[0]-1] + str2
            pos = (pos[0]-1,pos[1])
        elif path[pos] == "d":
            str1 = v[pos[1]-1] + str1
            str2 = "-" + str2
            pos = (pos[0],pos[1]-1)
        # print(pos)
            
    return Score, str1,str2           

v = input()
w = input()
n1 = len(v)
n2 = len(w)
Score,str1,str2 = FittingAlignment(v,w,n1,n2)
print(Score)
print(str1)
print(str2)
```
